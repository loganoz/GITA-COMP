## Gráfica de una función de varias variables 

En el siguiente ejemplo, creamos una representación de la gráfica de la función $f(x,y)=x^2+y^2$

```python
# Example of a 2-variable function and plotting it
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def f(x, y):
    return x**2 + y**2

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.title('Function of Two Variables')
plt.show()
```

## Conjuntos de nivel de la función
Ahora representamos los conjuntos de nivel de la función

```python
# Plotting level curves for a function of two variables
def g(x, y):
    return x**2 - y**2

Z = g(X, Y)

plt.contour(X, Y, Z, levels=20, cmap='plasma')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Level Curves of g(x, y)')
plt.colorbar(label='Value')
plt.show()
```

## Método de las secciones

Ejemplo de aplicación del método de la secciones (corte con planos) a la función $f(x,y)$

```python
# Example of applying the method of sections
# Consider the function f(x, y) = x^2 + y^2
# For a section at y = 2, find the resulting curve in 2D
y_section = 2
x_section = np.linspace(-5, 5, 100)
plt.plot(x_section, f(x_section, y_section))
plt.xlabel('X-axis')
plt.ylabel('f(x, 2)')
plt.title('Section at y = 2')
plt.show()
```
